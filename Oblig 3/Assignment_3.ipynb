```python
import csv
import pandas as pd
import sklearn as sk
from sklearn.model_selection import train_test_split
import numpy as np
import time
from sklearn.preprocessing import MinMaxScaler
from sklearn.neighbors import KNeighborsClassifier
```

```python
data = pd.read_csv('titanic.csv')
```

```python
data
```

```python
y = data.Survived
x = data.drop('Survived', axis=1)
print(x.columns)
```

```python
df = data.copy(deep=True)
for i in range(len(df.Sex)):
  if df.Sex[i] == 'male':
    df.Sex[i] = 0
  elif df.Sex[i] == 'female':
    df.Sex[i] = 1
  else:
    print('Accepted gender value not found')
    continue
    
df = df.drop('Name', axis=1)
```

```python
df
```

```python
scaler = MinMaxScaler()
data1 = scaler.fit_transform(df)
new_data = pd.DataFrame(data1)

new_data
```

```python
new_new_data = new_data.drop(0, axis=1)
X,y = new_new_data, new_data[0]

X_train, X_test, y_train, y_test = train_test_split(X, y)
print(X_train)
```

```python
neigh = KNeighborsClassifier(n_neighbors=3)
neigh.fit(X, y)
```

```python
score = (neigh.score(X_test, y_test)*100)
score_rounded = "{:.2f}".format(score)
print("The accuracy of K-nearest neighbor is: ", score_rounded, "%")
```
